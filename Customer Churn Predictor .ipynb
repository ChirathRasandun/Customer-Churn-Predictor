# Customer Churn Prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.CSV')
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data['TotalCharges'] = data['TotalCharges'].str.strip()

data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
data.info()
```

```python
data.isna().sum()
```

```python
data['TotalCharges'] =data['TotalCharges'].interpolate()
```

```python
data.isna().sum()
```

```python
new_data=data.drop('customerID',axis=1)
new_data.head()
```

```python
new_data['MultipleLines'].value_counts()
```

```python
new_data['InternetService'].value_counts()
```

```python
new_data['Contract'].value_counts()
```

```python
new_data['PaymentMethod'].value_counts()
```

```python
label_encoder = LabelEncoder()
new_data["Churn"] = label_encoder.fit_transform(new_data["Churn"])
new_data.head()
new_data.info()
```

```python
new_data=pd.get_dummies(new_data)
new_data.head()
```

```python
X = new_data.drop('Churn',axis=1)
y=new_data['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X_test.shape,X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
predicted=model.predict(X_test)
actual=y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(actual,predicted)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(actual,predicted)
```

```python
import pandas as pd
from sklearn.metrics import classification_report
report=pd.DataFrame(classification_report(actual,predicted,output_dict=True))
report
```
